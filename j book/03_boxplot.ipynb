```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti = sns.load_dataset('titanic')
kashti.head()
```

```python
sns.boxplot(x='survived', y='age' 
                        , showmeans=True
                        , meanprops={'marker': '*'
                                    ,'markersize': '15'
                                    ,'markeredgecolor': 'red'}
                                     
                        , data=kashti)
```

```python
sns.boxplot(x='survived', y='age' 
                        , showmeans=True
                        , meanprops={'marker': '*'
                                    ,'markersize': '15'
                                    ,'markeredgecolor': 'red'}        
                        , data=kashti)
#Show Labels
plt.xlabel('How many survived' , size=14 )
plt.ylabel('Age (years)', size=14 )
plt.title('Titanict World', size=18 , weight='bold' )
plt.show()
```
